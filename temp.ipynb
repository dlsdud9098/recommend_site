```python
import requests
from bs4 import BeautifulSoup

rq = requests.get('https://series.naver.com/novel/detail.series?productNo=12481255')
soup = BeautifulSoup(rq.text, 'html.parser')

img = soup.select_one('#container > div.aside.NE\=a\:nvi > span > img')
            
if img is None:
    print('이미지 없음')
    
    img = soup.select_one('#container > div > a > img').get('src')
else:
    img = img.get('src')

title = soup.select_one('#content > div.end_head > h2').get_text()
rating = soup.select_one('#content > div.end_head > div.score_area > em').get_text()
genre = soup.select_one('#content > ul.end_info.NE\=a\:nvi > li > ul > li:nth-child(2) > span > a').get_text()
serial = soup.select_one('#content > ul.end_info.NE\=a\:nvi > li > ul > li.ing > span').get_text()
publisher = soup.select_one('#content > ul.end_info.NE\=a\:nvi > li > ul > li:nth-child(4) > a').get_text()
summary = soup.select_one('#content > div.end_dsc > div').get_text()
page_count = soup.select_one('#content > h5 > strong').get_text()
page_unit = soup.select_one('#content > h5').get_text().strip()[-1]

novel_data = {
    'img': img,
    'title': title,
    'rating': rating,
    'genre': genre,
    'serial': serial,
    'publisher': publisher,
    'summary': summary,
    'page_count': page_count,
    'page_unit': page_unit
}

novel_data
```

```python
import pickle

with open('data/naver_novel_page_link_data.data', 'rb') as f:
    a = pickle.load(f)

# 데이터 나누기
def split_data(data, split_num):
    """리스트와 딕셔너리 모두 처리하는 범용 분할 함수"""
    
    if isinstance(data, list):
        # 리스트인 경우 (기존 로직)
        new_data = []
        for i in range(0, len(data), split_num):
            new_data.append(data[i: i+split_num])
        return new_data
    
    elif isinstance(data, dict):
        # 딕셔너리인 경우
        items = list(data.items())  # (key, value) 튜플들의 리스트
        new_data = []
        
        for i in range(0, len(items), split_num):
            batch_items = items[i: i+split_num]
            batch_dict = dict(batch_items)  # 다시 딕셔너리로 변환
            new_data.append(batch_dict)
        
        return new_data
    
    else:
        raise TypeError(f"지원하지 않는 타입입니다: {type(data)}")

# urls = split_data(a, 5)
a[0][0]
```
